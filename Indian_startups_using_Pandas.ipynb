#lets check the no. of startups according to year that made it to top 300

```python
data.head(2)
```

```python
data['Starting Year'].value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.bar(data['Starting Year'].value_counts().index,data['Starting Year'].value_counts().values)
plt.title("No. of startups according to year", fontdict={"fontsize":15})
plt.xlabel("Years",fontdict={"fontsize":15,"color":"b"})
plt.ylabel("No. of startups",fontdict={"fontsize":15,"color":"b"})
```

# lets check the city with highest no. of startups

```python
data.head(2)
```

```python
plot1=data["City"].value_counts().head(10)
plot1
```

```python
plt.plot(plot1,color="b",marker="*",markeredgecolor="r",linestyle=":") #ploting the data using matplotlib library
plt.xlabel("City",c="b") #labelling the data
plt.xticks(rotation='vertical')
plt.ylabel("No. of Startups",c="b")
plt.title("Cities with highest no. of startups", color= "g",size=15) #providing a title to the plot
plt.show()
```

# lets find the companies with highest funding rounds

```python
data.head(2)
```

```python
p1=data["Funding Round"].sort_values(ascending=False).head(10)
p1.index
```

```python
p2=data.loc[[11, 41, 113, 107, 50, 181, 80, 23, 239, 8],'Company']
```

```python
plt.figure(figsize=(8,6))
plt.barh(p2,p1,alpha=0.8)  
plt.title('Companies with highest funding rounds',fontsize=40)    #Note labelling the data
plt.ylabel('Company',fontsize=20)               #Note labelling the y-label
plt.xlabel('Count',fontsize=20)               #Note labelling the x-label
plt.tick_params(labelsize=8)
plt.show()
```

# lets find the top 10 companies with hiegest investors

```python
data.head(2)
```

```python
a1=data["No. of Investors"].sort_values(ascending=False).head(10)
a1    
```

```python
a2=data.loc[[11, 23, 50, 10, 32, 69, 31, 88, 4, 13],"Company"]
a2
```

```python
sns.barplot(x=a2,y=a1,data=data)
plt.xticks(rotation='vertical')
plt.title('Top 10 Companies with highest no. of Investors',size=20)
plt.xlabel('Company',size=15)
plt.ylabel('No. of Investors',size=15)
plt.tick_params(labelsize='x-small')
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(data.corr(),annot=True)
plt.savefig("heatmap.png", dpi=300)
```
